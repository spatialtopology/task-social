```python
def list_without_dsstore(sub_dir):
    filenames= os.listdir(sub_dir) # get all files' and folders' names in the current directory
    filename_result = []

    for filename in filenames: # loop through all the files and folders
        if os.path.isdir(os.path.join(sub_dir, filename)): # check whether the current object is a folder or not
            filename_result.append(filename)
    filename_result.sort()
    
    if '.DS_Store' in filename_result:
        filename_result.remove('.DS_Store')
    return filename_result
```
